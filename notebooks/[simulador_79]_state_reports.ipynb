```python
# Rio de Janeiro
# http://painel.saude.rj.gov.br/monitoramento/covid19.html

def compute_rj_data():
    return (pd.read_csv('../data/BICOVIDREAIS_26maio.csv')
              .query("classificacao == 'CONFIRMADO'")
              .pipe(lambda df: {
                  'state': 'RJ',
                  'deaths_young': len(df.query("(evolucao == 'OBITO') and (idade < 20)")),
                  'deaths_adult': len(df.query("(evolucao == 'OBITO') and (20 <= idade < 60)")),
                  'deaths_elder': len(df.query("(evolucao == 'OBITO') and (idade >= 60)")),
                  'deaths_total': len(df.query("evolucao == 'OBITO'")),
                  'cases_young': len(df.query("(idade < 20)")), 
                  'cases_adult': len(df.query("(20 <= idade < 60)")),                           
                  'cases_elder': len(df.query("(idade >= 60)")),
                  'cases_total': len(df)   
                  })
           )
rj_data = compute_rj_data()
states_data.append(rj_data)
```

```python
def compute_lethality(df):
    return df.assign(**{f"{subject}_lethality": (df[f"deaths_{subject}"]/df[f"cases_{subject}"])
                       for subject in ['total', 'adult', 'elder', 'young']})
```

```python
state_reports = pd.DataFrame(states_data).pipe(compute_lethality)
```

```python
state_reports = state_reports.append((state_reports.drop(columns=['state', 'total_lethality', 'adult_lethality', 'elder_lethality'])
              .sum()
              .to_frame()
              .T
              .assign(state='Brasil')
              .pipe(compute_lethality)), ignore_index=True)
```

```python
state_reports.to_csv('lethality_rates.csv', index=False)
```
